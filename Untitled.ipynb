```python
import os
import cv2
import numpy as np
from PIL import Image
from sklearn.model_selection import train_test_split
from keras.utils import normalize
from keras.models import Sequential
from keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout
from sklearn.metrics import confusion_matrix, accuracy_score
import matplotlib.pyplot as plt
import seaborn as sns

# Load Dataset
dataset = []
label = []
img_dir = './dataset/'

# Process "No Tumor" Images
no_tumor_img = os.listdir(img_dir + 'no/')
for img_name in no_tumor_img:
    if img_name.endswith('.jpg'):
        img = cv2.imread(img_dir + 'no/' + img_name)
        img = Image.fromarray(img, 'RGB')
        img = img.resize((128, 128))
        dataset.append(np.array(img))
        label.append(0)

# Process "Yes Tumor" Images
yes_tumor_img = os.listdir(img_dir + 'yes/')
for img_name in yes_tumor_img:
    if img_name.endswith('.jpg'):
        img = cv2.imread(img_dir + 'yes/' + img_name)
        img = Image.fromarray(img, 'RGB')
        img = img.resize((128, 128))
        dataset.append(np.array(img))
        label.append(1)

# Convert to numpy arrays
dataset = np.array(dataset)
label = np.array(label)

# Train-Test Split
x_train, x_test, y_train, y_test = train_test_split(dataset, label, test_size=0.2, random_state=42)

# Normalize the data
x_train = normalize(x_train, axis=1)
x_test = normalize(x_test, axis=1)

# Define the CNN Model
cnn_model = Sequential([
    Conv2D(32, (3, 3), activation='relu', input_shape=(128, 128, 3)),
    MaxPooling2D(pool_size=(2, 2)),
    Conv2D(64, (3, 3), activation='relu'),
    MaxPooling2D(pool_size=(2, 2)),
    Conv2D(128, (3, 3), activation='relu'),
    MaxPooling2D(pool_size=(2, 2)),
    Flatten(),
    Dense(128, activation='relu'),
    Dropout(0.5),
    Dense(1, activation='sigmoid')  # Output layer for binary classification
])

# Compile the Model
cnn_model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# Train the Model
cnn_history = cnn_model.fit(x_train, y_train, epochs=10, validation_data=(x_test, y_test), batch_size=16, verbose=1)

# Evaluate the Model
res = cnn_model.evaluate(x_test, y_test, verbose=0)
print(f"Test Accuracy: {res[1]*100:.2f}%\nTest Loss: {res[0]:.4f}")

# Predict on Test Data
y_pred = cnn_model.predict(x_test)
y_pred_classes = np.where(y_pred > 0.5, 1, 0)

# Output Predictions
for i in range(len(y_pred_classes)):
    result = "Yes, tumor detected" if y_pred_classes[i] == 1 else "No tumor detected"
    print(f"Sample {i+1}: {result}")

# Confusion Matrix
cm = confusion_matrix(y_test, y_pred_classes)

# Plot Confusion Matrix
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", xticklabels=['No Tumor', 'Yes Tumor'], yticklabels=['No Tumor', 'Yes Tumor'])
plt.title("Confusion Matrix")
plt.xlabel("Predicted Labels")
plt.ylabel("True Labels")
plt.show()
```
